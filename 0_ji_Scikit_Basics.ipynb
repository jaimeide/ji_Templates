### Classification

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report


# Load data
df = pd.read_csv('0_data/titanic_train.csv')
df.head()

## PREPROCESSING

# Show missing values
ax = sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis') 
ax.set_title('Missing values')
plt.show()

# Look at the class to predict
# Counting the survivors
sns.set_style('whitegrid')
ax = sns.countplot(x='Survived',data=df,palette='RdBu_r')
ax.set_title('Survived')

# Replace Age with mean value
df['Age'].fillna((df['Age'].mean()), inplace=True)
#display(df)

# Dropout feature with too much missing value
df = df.drop('Cabin',axis=1)

# Convert categorical to number using dummy
sex = pd.get_dummies(df['Sex'],drop_first=True) # notice that these are binary categories. So we dropout the first column

# Drop and merge converted feats
aux = df.drop(['Name','Sex','Embarked','Ticket'],axis=1)

# Concatenate the discretized features
df = pd.concat([aux,sex],axis=1)
display(df)

## MODEL FIT

# Train and test split
X_train,X_test, y_train,y_test = train_test_split(df.drop(['Survived'],axis=1),
                                                  df['Survived'],test_size=0.3,random_state=101)
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)

# Define and fit model
#model = LogisticRegression(penalty='l2',C=1.0)
model = LogisticRegression()
model.fit(X_train,y_train)

# Predict and Evaluate
y_pred = model.predict(X_test)
print(classification_report(y_test,y_pred))
```

## 2.3) Feat Engineering

```python
# Fill-in missing age with Pclass as proxy
def fillin_age(cols):
    age = cols[0]
    Pclass= cols[1]
    # function that given an age, return a value based on its Pclass
    if pd.isnull(age): # only if is NaN
        if Pclass==1:
            return 37
        if Pclass==2:
            return 29
        if Pclass == 3:
            return 24
    else:
        return age
    
# Using apply with DF columns
df_train['Age'] = df_train[['Age','Pclass']].apply(fillin_age,axis=1)
```

```python
# Dropout feature with too much missing value
df_train = df_train.drop('Cabin',axis=1)
df_train
```

```python
# Convert CATEGORY --> NUMBERS

# Examine DF
df_train.info()
```

```python
# CONVERTION using dummy
sex = pd.get_dummies(df_train['Sex'],drop_first=True) # notice that these are binary categories. So we dropout the first column
embarked = pd.get_dummies(df_train['Embarked'],drop_first=True) # Actually, first col is redundant since they are mutually exclusive!
embarked
```

```python
aux = df_train.drop(['Name','Sex','Embarked','Ticket'],axis=1)
aux
```

```python
# Concatenate the discretized features
df_train = pd.concat([aux,sex,embarked],axis=1)
df_train
```

## 3.4) Train and test

```python
from sklearn.model_selection import train_test_split

X_train,X_test, y_train,y_test = train_test_split(df_train.drop(['Survived'],axis=1),
                                                  df_train['Survived'],test_size=0.3,random_state=101)
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

## 3.5) Define Model+ Fit

```python
from sklearn.linear_model import LogisticRegression

#model = LogisticRegression(penalty='l2',C=1.0)
model = LogisticRegression()
model.fit(X_train,y_train)
```

## 3.6) Test + Validate

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)
print(classification_report(y_test,y_pred))
```
